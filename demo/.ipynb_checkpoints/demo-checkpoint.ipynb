## Test data

In this section, we demonstrate how to evaluate the pretrained EfficientNet-B0 model on a new test dataset. The test data is organized into folders for each class and placed in the `demo/data` directory. 

#### Steps:
1. **Data Preprocessing**: The test data is preprocessed to match the input requirements of EfficientNet-B0. This includes resizing to 224x224 and applying standard normalization.
2. **Model Loading**: We load the pretrained EfficientNet-B0 model and replace its classifier to match the number of classes in our dataset.
3. **Checkpoint Restoration**: The model weights are restored from the checkpoint `checkpoints/efficientnet_model_epoch_8.pth`.
4. **Evaluation**: The model is evaluated on the test dataset, and we calculate:
   - Overall test accuracy
   - Per-class accuracies
5. **Output**: The results include the overall accuracy of the model across all test data and accuracy for each individual class.

This script allows users to quickly verify the performance of the model on unseen data.

```python
import torch
import torch.nn as nn
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from torchvision.models import efficientnet_b0, EfficientNet_B0_Weights
import os

# Paths
model_path = "../checkpoints/efficientnet_model_epoch_8.pth"
test_data_path = "data"

# Data preprocessing
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])  # Standard normalization
])

# Load test dataset
test_dataset = datasets.ImageFolder(test_data_path, transform=transform)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Load the pretrained EfficientNet-B0 model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = efficientnet_b0(weights=EfficientNet_B0_Weights.IMAGENET1K_V1)

# Replace the classifier to match the number of classes
num_classes = len(test_dataset.classes)
model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)
model = model.to(device)

# Load model checkpoint
checkpoint = torch.load(model_path, map_location=device)
model.load_state_dict(checkpoint['model_state_dict'])
model.eval()

# Evaluation function
def evaluate_model(model, dataloader):
    correct = 0
    total = 0
    class_correct = [0] * num_classes
    class_total = [0] * num_classes

    with torch.no_grad():
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            for label, prediction in zip(labels, predicted):
                if label == prediction:
                    class_correct[label] += 1
                class_total[label] += 1

    overall_accuracy = 100 * correct / total
    class_accuracies = [100 * (c / t if t > 0 else 0) for c, t in zip(class_correct, class_total)]

    return overall_accuracy, class_accuracies

# Evaluate the model
overall_accuracy, class_accuracies = evaluate_model(model, test_loader)

# Print results
print(f"Overall Test Accuracy: {overall_accuracy:.2f}%")
for i, class_acc in enumerate(class_accuracies):
    print(f"Class {test_dataset.classes[i]} Accuracy: {class_acc:.2f}%")
```

### Random 10 correct predictions

```python
import matplotlib.pyplot as plt
import random

# Function to plot 10 correct predictions
def plot_correct_predictions(model, dataloader, class_names):
    correct_images = []
    correct_labels = []

    model.eval()
    with torch.no_grad():
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)

            # Collect correct predictions
            for i in range(len(labels)):
                if len(correct_images) >= 10:  # Stop after 10 correct predictions
                    break
                if labels[i] == predicted[i]:
                    correct_images.append(images[i].cpu())
                    correct_labels.append(labels[i].cpu())

    # Randomly select 10 correct predictions
    indices = random.sample(range(len(correct_images)), min(10, len(correct_images)))

    # Plot the images
    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    axes = axes.flatten()

    for idx, ax in enumerate(axes):
        if idx < len(indices):
            img = correct_images[indices[idx]].permute(1, 2, 0).numpy()
            img = img * [0.229, 0.224, 0.225] + [0.485, 0.456, 0.406]  # Denormalize
            img = img.clip(0, 1)

            ax.imshow(img)
            ax.axis('off')
            ax.set_title(f"Class: {class_names[correct_labels[indices[idx]].item()]}")
        else:
            ax.axis('off')

    plt.tight_layout()
    plt.show()

# Plot 10 correct predictions
plot_correct_predictions(model, test_loader, test_dataset.classes)
```

### Random 10 mispredictions

```python
import matplotlib.pyplot as plt
import random

# Function to plot 10 mispredictions
def plot_mispredictions(model, dataloader, class_names):
    mispredicted_images = []
    actual_labels = []
    predicted_labels = []

    model.eval()
    with torch.no_grad():
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)

            # Collect mispredictions
            for i in range(len(labels)):
                if labels[i] != predicted[i]:
                    mispredicted_images.append(images[i].cpu())
                    actual_labels.append(labels[i].cpu())
                    predicted_labels.append(predicted[i].cpu())

    # Randomly select 10 mispredictions
    indices = random.sample(range(len(mispredicted_images)), min(10, len(mispredicted_images)))

    # Plot the images
    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    axes = axes.flatten()

    for idx, ax in enumerate(axes):
        if idx < len(indices):
            img = mispredicted_images[indices[idx]].permute(1, 2, 0).numpy()
            img = img * [0.229, 0.224, 0.225] + [0.485, 0.456, 0.406]  # Denormalize
            img = img.clip(0, 1)

            ax.imshow(img)
            ax.axis('off')
            ax.set_title(f"Actual: {class_names[actual_labels[indices[idx]].item()]}\nPred: {class_names[predicted_labels[indices[idx]].item()]}")
        else:
            ax.axis('off')

    plt.tight_layout()
    plt.show()

# Plot 10 mispredictions
plot_mispredictions(model, test_loader, test_dataset.classes)
```
